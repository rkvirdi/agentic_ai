```python
t=read_files('../data_file')
text_chunks=chunk_text(t)
```

```python
chunk=[]
for t in text_chunks:
    for k in t:
        chunk.append(k)
chunk
```

```python
from langchain_chroma import Chroma
from utils.embeddings import sentence_transformer_embeding_model
```

```python
embedding=sentence_transformer_embeding_model()
vectorstore=Chroma.from_documents(chunk,embedding,persist_directory="../newchromadb_file",collection_name="oscars")
docs=vectorstore.similarity_search("who won best actor award")
docs
```
